```python
# ====================
# 1. KONFIGURASI & SETUP
# ====================

import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
import seaborn as sns
from dataclasses import dataclass, field
from typing import Dict, List, Tuple, Optional
import math

# Setup plotting
plt.style.use('default')
sns.set_style("whitegrid")

@dataclass
class TankConfig:
    """Konfigurasi parameter water tank"""
    
    # Parameter geometri tank (silinder vertikal)
    tank_height: float = 3.0          # m
    tank_diameter: float = 2.0         # m
    tank_area: float = field(init=False, default=None)  # m²
    
    # Parameter pipa
    inlet_diameter: float = 0.1        # m
    inlet_area: float = field(init=False, default=None)  # m²
    outlet_diameter: float = 0.1       # m
    outlet_area: float = field(init=False, default=None) # m²
    
    # Parameter aliran
    inlet_velocity: float = 1.5        # m/s (kecepatan aliran masuk)
    outlet_velocity: float = 1.5       # m/s (kecepatan aliran keluar)
    
    # Parameter fisik
    initial_height: float = 0.5         # m (ketinggian awal air)
    g: float = 9.81                     # m/s² (gravitasi)
    rho_water: float = 1000.0           # kg/m³ (densitas air)
    
    # Koefisien losses
    discharge_coeff: float = 0.6         # koefisien discharge untuk orifice
    
    # Parameter kontrol
    pump_on: bool = True                 # status pompa inlet
    valve_open: bool = True              # status valve outlet
    
    # Parameter simulasi
    simulation_time: float = 300.0       # detik
    time_step: float = 1.0                # detik
    
    def __post_init__(self):
        """Validasi konfigurasi dan hitung atribut turunan"""
        self.tank_area = math.pi * (self.tank_diameter/2)**2
        self.inlet_area = math.pi * (self.inlet_diameter/2)**2
        self.outlet_area = math.pi * (self.outlet_diameter/2)**2
        self.max_volume = self.tank_area * self.tank_height
        
        if self.initial_height > self.tank_height:
            print("Peringatan: Ketinggian awal melebihi tinggi tank")
        if self.initial_height < 0:
            print("Peringatan: Ketinggian awal tidak boleh negatif")
    
    def copy(self):
        """Buat salinan konfigurasi"""
        params = {k: v for k, v in self.__dict__.items() 
                 if k not in ['tank_area', 'inlet_area', 'outlet_area', 'max_volume']}
        new_config = TankConfig(**params)
        return new_config
    
    def update_parameter(self, parameter_name: str, value: float):
        """Update satu parameter dan hitung ulang atribut turunan"""
        if parameter_name in self.__annotations__:
            setattr(self, parameter_name, value)
            self.__post_init__()
        else:
            raise ValueError(f"Parameter {parameter_name} tidak valid")

print("KONFIGURASI & SETUP")
```

```python
# ====================
# 2. MODEL FISIKA
# ====================

class WaterTankPhysics:
    """Model fisika untuk water tank"""
    
    def __init__(self, config: TankConfig):
        self.config = config
    
    def inlet_flow_rate(self, water_height: float) -> float:
        """Hitung laju aliran masuk (m³/s)"""
        if not self.config.pump_on:
            return 0.0
            
        if water_height >= self.config.tank_height - 0.01:
            return 0.0  # Tank hampir penuh, hentikan pengisian
            
        # Q_in = A_inlet * v_inlet
        return self.config.inlet_area * self.config.inlet_velocity
    
    def outlet_flow_rate(self, water_height: float) -> float:
        """Hitung laju aliran keluar berdasarkan ketinggian air (m³/s)"""
        if not self.config.valve_open:
            return 0.0
            
        if water_height <= 0.01:
            return 0.0  # Tank hampir kosong
            
        # Model Torricelli: v = Cd * sqrt(2gh)
        # Kecepatan keluar berdasarkan tekanan hidrostatik
        outlet_velocity = self.config.discharge_coeff * np.sqrt(2 * self.config.g * max(0, water_height))
        
        return self.config.outlet_area * outlet_velocity
    
    def net_flow_rate(self, water_height: float) -> float:
        """Hitung laju aliran neto (m³/s)"""
        Q_in = self.inlet_flow_rate(water_height)
        Q_out = self.outlet_flow_rate(water_height)
        return Q_in - Q_out
    
    def volume_to_height(self, volume: float) -> float:
        """Konversi volume ke ketinggian"""
        return volume / self.config.tank_area
    
    def height_to_volume(self, height: float) -> float:
        """Konversi ketinggian ke volume"""
        return height * self.config.tank_area
    
    def water_mass(self, water_height: float) -> float:
        """Hitung massa air dalam tank (kg)"""
        volume = self.height_to_volume(water_height)
        return volume * self.config.rho_water

print("MODEL FISIKA")
```

```python
# ====================
# 3. SISTEM PERSAMAAN DIFERENSIAL
# ====================

class TankDifferentialEquations:
    """Sistem persamaan diferensial untuk simulasi kontinu water tank"""
    
    def __init__(self, physics_model: WaterTankPhysics):
        self.physics = physics_model
        self.config = physics_model.config
    
    def system_equations(self, t: float, y: np.ndarray) -> np.ndarray:
        """
        Sistem persamaan diferensial:
        y = [water_height]
        
        Returns:
        dy/dt = [dh/dt]
        """
        h = y[0]  # ketinggian air
        
        # Batasi ketinggian antara 0 dan tinggi maksimum
        h = np.clip(h, 0, self.config.tank_height)
        
        # Hitung laju aliran neto
        Q_net = self.physics.net_flow_rate(h)
        
        # Perubahan ketinggian: dh/dt = Q_net / A_tank
        if self.config.tank_area > 0:
            dh_dt = Q_net / self.config.tank_area
        else:
            dh_dt = 0.0
            
        # Jika ketinggian mencapai batas, set laju perubahan ke 0
        if (h >= self.config.tank_height - 0.01 and dh_dt > 0) or (h <= 0.01 and dh_dt < 0):
            dh_dt = 0.0
            
        return np.array([dh_dt])
    
    def get_initial_conditions(self) -> np.ndarray:
        """Kondisi awal sistem"""
        return np.array([self.config.initial_height])

print("SISTEM PERSAMAAN DIFERENSIAL")
```

```python
# ====================
# 4. SIMULATOR UTAMA (DIPERBAIKI)
# ====================

class WaterTankSimulator:
    """Simulator utama water tank"""
    
    def __init__(self, config: TankConfig):
        self.config = config
        self.physics = WaterTankPhysics(config)
        self.equations = TankDifferentialEquations(self.physics)
        
        # Results storage
        self.time_history = None
        self.height_history = None
        self.volume_history = None
        self.inflow_history = None
        self.outflow_history = None
        self.netflow_history = None
        self.results = None
    
    def run_simulation(self) -> Dict:
        """Jalankan simulasi"""
        print("Memulai simulasi water tank...")
        
        # Setup time
        t_span = (0, self.config.simulation_time)
        t_eval = np.arange(0, self.config.simulation_time, self.config.time_step)
        
        # Initial conditions
        y0 = self.equations.get_initial_conditions()
        
        # Solve ODE system
        solution = solve_ivp(
            fun=self.equations.system_equations,
            t_span=t_span,
            y0=y0,
            t_eval=t_eval,
            method='RK45',
            rtol=1e-6,
            atol=1e-9,
            dense_output=True
        )
        
        # Store results
        self.time_history = solution.t
        self.height_history = solution.y[0]
        self.volume_history = self.height_history * self.config.tank_area
        
        # Calculate flow histories
        self.inflow_history = np.zeros_like(self.time_history)
        self.outflow_history = np.zeros_like(self.time_history)
        self.netflow_history = np.zeros_like(self.time_history)
        
        for i, h in enumerate(self.height_history):
            self.inflow_history[i] = self.physics.inlet_flow_rate(h)
            self.outflow_history[i] = self.physics.outlet_flow_rate(h)
            self.netflow_history[i] = self.inflow_history[i] - self.outflow_history[i]
        
        # Calculate metrics
        self.results = self._calculate_metrics()
        
        print("Simulasi selesai!")
        return self.results
    
    def _calculate_metrics(self) -> Dict:
        """Hitung metrik kinerja water tank"""
        if self.time_history is None:
            raise ValueError("Jalankan simulasi terlebih dahulu")
        
        # PERBAIKAN: ganti np.trapz dengan np.trapezoid
        metrics = {
            # Time metrics
            'time_to_empty': self._get_time_to_height(0.05),
            'time_to_full': self._get_time_to_height(self.config.tank_height - 0.05),
            'time_to_half': self._get_time_to_height(self.config.tank_height/2),
            
            # Height metrics
            'max_height': float(np.max(self.height_history)),
            'min_height': float(np.min(self.height_history)),
            'final_height': float(self.height_history[-1]),
            'avg_height': float(np.mean(self.height_history)),
            
            # Volume metrics
            'max_volume': float(np.max(self.volume_history)),
            'min_volume': float(np.min(self.volume_history)),
            'final_volume': float(self.volume_history[-1]),
            
            # Flow metrics (konversi ke m³/jam untuk readability)
            'max_inflow': float(np.max(self.inflow_history) * 3600),
            'max_outflow': float(np.max(self.outflow_history) * 3600),
            'avg_inflow': float(np.mean(self.inflow_history) * 3600),
            'avg_outflow': float(np.mean(self.outflow_history) * 3600),
            
            # Total volume (integral) - PERBAIKAN DISINI
            'total_inflow': float(np.trapezoid(self.inflow_history, self.time_history)),
            'total_outflow': float(np.trapezoid(self.outflow_history, self.time_history)),
            'net_volume_change': float(self.volume_history[-1] - self.volume_history[0]),
        }
        
        return metrics
    
    def _get_time_to_height(self, target_height: float) -> Optional[float]:
        """Waktu untuk mencapai ketinggian tertentu"""
        if target_height < 0 or target_height > self.config.tank_height:
            return None
            
        # Cari indeks pertama yang mencapai target
        if len(self.height_history) == 0:
            return None
            
        if self.height_history[0] <= target_height:
            # Pengisian (naik)
            indices = np.where(self.height_history >= target_height)[0]
        else:
            # Pengosongan (turun)
            indices = np.where(self.height_history <= target_height)[0]
            
        if len(indices) > 0:
            return float(self.time_history[indices[0]])
        return None

print("SIMULATOR UTAMA (DIPERBAIKI)")
```

```python
# ====================
# 5. VISUALISASI (Matplotlib) - DIPERBAIKI
# ====================

class Visualization:
    """Kelas untuk visualisasi hasil simulasi dengan matplotlib"""
    
    @staticmethod
    def plot_height_profile(simulator: WaterTankSimulator, ax: plt.Axes = None, title_prefix=""):
        """Plot profil ketinggian air"""
        if ax is None:
            fig, ax = plt.subplots(figsize=(10, 6))
        
        time = simulator.time_history
        height = simulator.height_history
        config = simulator.config
        
        ax.plot(time, height, 'b-', linewidth=2.5, label='Ketinggian Air', alpha=0.8)
        
        # Reference lines
        ax.axhline(y=config.tank_height, color='r', linestyle='--', 
                  linewidth=1.5, alpha=0.7, label=f'Tinggi Maks ({config.tank_height:.1f} m)')
        ax.axhline(y=0, color='g', linestyle='--', 
                  linewidth=1.5, alpha=0.7, label='Tank Kosong')
        
        # Fill area
        ax.fill_between(time, 0, height, alpha=0.2, color='blue', label='Volume Air')
        
        ax.set_xlabel('Waktu (detik)', fontsize=12)
        ax.set_ylabel('Ketinggian Air (m)', fontsize=12)
        if title_prefix:
            ax.set_title(f'{title_prefix} - Profil Ketinggian Air', fontsize=14, fontweight='bold')
        else:
            ax.set_title('Profil Ketinggian Air dalam Tank', fontsize=14, fontweight='bold')
        ax.legend(loc='best', fontsize=10)
        ax.grid(True, alpha=0.3)
        
        return ax
    
    @staticmethod
    def plot_flow_rates(simulator: WaterTankSimulator):
        """Plot laju aliran"""
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
        
        time = simulator.time_history
        inflow = simulator.inflow_history * 3600  # m³/h
        outflow = simulator.outflow_history * 3600  # m³/h
        netflow = simulator.netflow_history * 3600  # m³/h
        
        # Plot 1: Individual flow rates
        ax1.plot(time, inflow, 'g-', linewidth=2, label='Aliran Masuk', alpha=0.8)
        ax1.plot(time, outflow, 'r-', linewidth=2, label='Aliran Keluar', alpha=0.8)
        ax1.set_xlabel('Waktu (detik)', fontsize=12)
        ax1.set_ylabel('Laju Aliran (m³/jam)', fontsize=12)
        ax1.set_title('Laju Aliran Masuk dan Keluar', fontsize=14, fontweight='bold')
        ax1.legend(loc='best', fontsize=10)
        ax1.grid(True, alpha=0.3)
        
        # Plot 2: Net flow rate
        ax2.plot(time, netflow, 'purple', linewidth=2.5, label='Aliran Neto', alpha=0.8)
        ax2.axhline(y=0, color='k', linestyle='-', linewidth=0.5, alpha=0.5)
        ax2.fill_between(time, 0, netflow, where=(netflow>0), color='green', alpha=0.3, label='Pengisian')
        ax2.fill_between(time, 0, netflow, where=(netflow<0), color='red', alpha=0.3, label='Pengosongan')
        ax2.set_xlabel('Waktu (detik)', fontsize=12)
        ax2.set_ylabel('Laju Aliran Neto (m³/jam)', fontsize=12)
        ax2.set_title('Laju Aliran Neto', fontsize=14, fontweight='bold')
        ax2.legend(loc='best', fontsize=10)
        ax2.grid(True, alpha=0.3)
        
        plt.tight_layout()
        return fig
    
    @staticmethod
    def plot_volume_profile(simulator: WaterTankSimulator):
        """Plot profil volume"""
        fig, ax = plt.subplots(figsize=(10, 6))
        
        time = simulator.time_history
        volume = simulator.volume_history
        config = simulator.config
        
        ax.plot(time, volume, 'purple', linewidth=2.5, label='Volume Air', alpha=0.8)
        ax.axhline(y=config.max_volume, color='r', linestyle='--', 
                  linewidth=1.5, alpha=0.7, label=f'Volume Maks ({config.max_volume:.2f} m³)')
        ax.fill_between(time, 0, volume, alpha=0.2, color='purple')
        
        ax.set_xlabel('Waktu (detik)', fontsize=12)
        ax.set_ylabel('Volume (m³)', fontsize=12)
        ax.set_title('Profil Volume Air dalam Tank', fontsize=14, fontweight='bold')
        ax.legend(loc='best', fontsize=10)
        ax.grid(True, alpha=0.3)
        
        return fig
    
    @staticmethod
    def plot_phase_diagram(simulator: WaterTankSimulator, ax: plt.Axes = None):
        """Plot diagram fase operasi tank"""
        if ax is None:
            fig, ax = plt.subplots(figsize=(10, 4))
        
        time = simulator.time_history
        height = simulator.height_history
        config = simulator.config
        
        # Define phases based on water level
        phases = []
        for h in height:
            if h <= 0.1 * config.tank_height:
                phases.append(0)  # Kosong
            elif h <= 0.3 * config.tank_height:
                phases.append(1)  # Rendah
            elif h <= 0.6 * config.tank_height:
                phases.append(2)  # Sedang
            elif h <= 0.9 * config.tank_height:
                phases.append(3)  # Tinggi
            else:
                phases.append(4)  # Hampir Penuh
        
        # Color mapping
        phase_colors = ['lightgray', 'lightblue', 'lightgreen', 'yellow', 'orange']
        phase_labels = ['Kosong', 'Rendah', 'Sedang', 'Tinggi', 'Hampir Penuh']
        
        # Plot phases
        current_phase = phases[0]
        start_idx = 0
        
        for i, phase in enumerate(phases[1:], 1):
            if phase != current_phase or i == len(phases) - 1:
                ax.fill_betweenx([0, 1], time[start_idx], time[i-1], 
                                color=phase_colors[current_phase], alpha=0.4)
                mid_time = (time[start_idx] + time[i-1]) / 2
                ax.text(mid_time, 0.5, phase_labels[current_phase], 
                       ha='center', va='center', fontweight='bold', fontsize=9)
                current_phase = phase
                start_idx = i
        
        ax.set_xlabel('Waktu (detik)', fontsize=12)
        ax.set_ylabel('Fase Operasi', fontsize=12)
        ax.set_title('Diagram Fase Operasi Tank', fontsize=14, fontweight='bold')
        ax.set_ylim([0, 1])
        ax.set_yticks([])
        ax.grid(True, alpha=0.3)
        
        return ax

print("VISUALISASI")
```

```python
# ====================
# 6. ANALISIS SKENARIO (DIPERBAIKI)
# ====================

class ScenarioAnalysis:
    """Analisis berbagai skenario water tank"""
    
    @staticmethod
    def analyze_filling_only(base_config: TankConfig):
        """Analisis skenario hanya pengisian"""
        config = base_config.copy()
        config.valve_open = False
        config.pump_on = True
        
        simulator = WaterTankSimulator(config)
        simulator.run_simulation()
        return simulator
    
    @staticmethod
    def analyze_emptying_only(base_config: TankConfig):
        """Analisis skenario hanya pengosongan"""
        config = base_config.copy()
        config.pump_on = False
        config.valve_open = True
        
        simulator = WaterTankSimulator(config)
        simulator.run_simulation()
        return simulator
    
    @staticmethod
    def analyze_simultaneous(base_config: TankConfig):
        """Analisis skenario pengisian dan pengosongan bersamaan"""
        config = base_config.copy()
        config.pump_on = True
        config.valve_open = True
        
        simulator = WaterTankSimulator(config)
        simulator.run_simulation()
        return simulator
    
    @staticmethod
    def analyze_pump_variations(base_config: TankConfig):
        """Analisis variasi kecepatan pompa"""
        variations = []
        velocities = [0.5, 1.0, 1.5, 2.0, 2.5]
        
        for v in velocities:
            config = base_config.copy()
            config.inlet_velocity = v
            config.valve_open = False  # Hanya pengisian
            
            simulator = WaterTankSimulator(config)
            simulator.run_simulation()
            variations.append({
                'velocity': v,
                'simulator': simulator,
                'metrics': simulator.results
            })
        
        return variations
    
    @staticmethod
    def analyze_valve_variations(base_config: TankConfig):
        """Analisis variasi bukaan valve"""
        variations = []
        discharge_coeffs = [0.3, 0.4, 0.5, 0.6, 0.7, 0.8]
        
        # Set initial height to full for emptying analysis
        config = base_config.copy()
        config.initial_height = config.tank_height
        config.pump_on = False
        
        for cd in discharge_coeffs:
            test_config = config.copy()
            test_config.discharge_coeff = cd
            
            simulator = WaterTankSimulator(test_config)
            simulator.run_simulation()
            variations.append({
                'discharge_coeff': cd,
                'simulator': simulator,
                'metrics': simulator.results
            })
        
        return variations
    
    @staticmethod
    def find_optimal_tank_size(daily_demand: float, peak_hours: int = 4):
        """
        Menentukan ukuran tank optimal berdasarkan kebutuhan air harian
        
        daily_demand: kebutuhan air harian (m³)
        peak_hours: jam puncak pemakaian (jam)
        """
        # Asumsi: tank harus mampu menampung kebutuhan selama peak hours
        # dengan faktor keamanan 1.5
        safety_factor = 1.5
        required_capacity = (daily_demand / 24) * peak_hours * safety_factor
        
        # Rekomendasi dimensi tank (silinder vertikal dengan rasio H:D = 1.5:1)
        # Volume = π * (D/2)² * H, dengan H = 1.5 * D
        # Volume = π * (D/2)² * 1.5D = 1.5π * D³/4
        optimal_diameter = (required_capacity * 4 / (1.5 * math.pi)) ** (1/3)
        optimal_height = 1.5 * optimal_diameter
        
        # Estimasi waktu pengisian dengan diameter pipa 0.1 m dan kecepatan 1.5 m/s
        pipe_area = math.pi * (0.1/2)**2
        filling_rate = pipe_area * 1.5  # m³/s
        
        return {
            'required_capacity': required_capacity,
            'recommended_height': optimal_height,
            'recommended_diameter': optimal_diameter,
            'recommended_area': math.pi * (optimal_diameter/2)**2,
            'filling_time_estimate': required_capacity / filling_rate if filling_rate > 0 else 0
        }

print("ANALISIS SKENARIO")
```

```python
# ====================
# 7. MENJALANKAN SIMULASI (DIPERBAIKI)
# ====================

def main():
    """Program utama"""
    print("="*60)
    print("SIMULASI WATER TANK SYSTEM")
    print("="*60)
    
    # 1. Setup konfigurasi dasar
    print("\n1. Setup Konfigurasi Dasar...")
    base_config = TankConfig(
        tank_height=3.0,
        tank_diameter=2.0,
        inlet_diameter=0.1,
        outlet_diameter=0.1,
        inlet_velocity=1.5,
        outlet_velocity=1.5,
        initial_height=0.5,
        simulation_time=300.0
    )
    
    print(f"   - Tinggi tank: {base_config.tank_height} m")
    print(f"   - Diameter tank: {base_config.tank_diameter} m")
    print(f"   - Luas tank: {base_config.tank_area:.2f} m²")
    print(f"   - Volume maks: {base_config.max_volume:.2f} m³")
    
    # 2. Jalankan simulasi untuk berbagai skenario
    print("\n2. Menjalankan Simulasi Skenario...")
    
    # Skenario 1: Hanya pengisian
    print("\n   Skenario 1: Hanya Pengisian (Pompa ON, Valve CLOSED)")
    filling_sim = ScenarioAnalysis.analyze_filling_only(base_config)
    
    # Skenario 2: Hanya pengosongan
    print("\n   Skenario 2: Hanya Pengosongan (Pompa OFF, Valve OPEN)")
    emptying_config = base_config.copy()
    emptying_config.initial_height = base_config.tank_height  # Mulai dari penuh
    emptying_sim = ScenarioAnalysis.analyze_emptying_only(emptying_config)
    
    # Skenario 3: Pengisian dan pengosongan bersamaan
    print("\n   Skenario 3: Pengisian & Pengosongan Bersamaan")
    simultaneous_sim = ScenarioAnalysis.analyze_simultaneous(base_config)
    
    # 3. Tampilkan hasil
    print("\n3. Hasil Simulasi:")
    print("-"*40)
    
    print("\n   Skenario 1 - Hanya Pengisian:")
    if filling_sim.results['time_to_full']:
        print(f"   - Waktu mencapai penuh: {filling_sim.results['time_to_full']:.1f} detik")
    else:
        print(f"   - Waktu mencapai penuh: Tidak tercapai dalam {base_config.simulation_time} detik")
    print(f"   - Volume akhir: {filling_sim.results['final_volume']:.2f} m³")
    print(f"   - Total air masuk: {filling_sim.results['total_inflow']:.2f} m³")
    
    print("\n   Skenario 2 - Hanya Pengosongan:")
    if emptying_sim.results['time_to_empty']:
        print(f"   - Waktu mencapai kosong: {emptying_sim.results['time_to_empty']:.1f} detik")
    else:
        print(f"   - Waktu mencapai kosong: Tidak tercapai dalam {base_config.simulation_time} detik")
    print(f"   - Volume akhir: {emptying_sim.results['final_volume']:.2f} m³")
    print(f"   - Total air keluar: {emptying_sim.results['total_outflow']:.2f} m³")
    
    print("\n   Skenario 3 - Bersamaan:")
    print(f"   - Ketinggian akhir: {simultaneous_sim.results['final_height']:.2f} m")
    print(f"   - Volume akhir: {simultaneous_sim.results['final_volume']:.2f} m³")
    print(f"   - Perubahan volume: {simultaneous_sim.results['net_volume_change']:.2f} m³")
    
    # 4. Analisis optimalisasi
    print("\n4. Analisis Optimalisasi Tank...")
    daily_demand = 50  # m³/hari (asumsi kebutuhan asrama)
    peak_hours = 4      # jam puncak
    
    optimal = ScenarioAnalysis.find_optimal_tank_size(daily_demand, peak_hours)
    print(f"\n   Berdasarkan kebutuhan {daily_demand} m³/hari dengan peak {peak_hours} jam:")
    print(f"   - Kapasitas diperlukan: {optimal['required_capacity']:.2f} m³")
    print(f"   - Tinggi rekomendasi: {optimal['recommended_height']:.2f} m")
    print(f"   - Diameter rekomendasi: {optimal['recommended_diameter']:.2f} m")
    print(f"   - Estimasi waktu pengisian: {optimal['filling_time_estimate']:.1f} detik")
    
    # 5. Visualisasi
    print("\n5. Membuat Visualisasi...")
    
    # Plot untuk ketiga skenario
    fig1, axes = plt.subplots(2, 2, figsize=(14, 10))
    
    Visualization.plot_height_profile(filling_sim, axes[0, 0], "Pengisian")
    Visualization.plot_height_profile(emptying_sim, axes[0, 1], "Pengosongan")
    Visualization.plot_height_profile(simultaneous_sim, axes[1, 0], "Bersamaan")
    Visualization.plot_phase_diagram(simultaneous_sim, axes[1, 1])
    
    plt.tight_layout()
    plt.show()
    
    # Plot flow rates untuk skenario bersamaan
    print("\n   Menampilkan grafik laju aliran (Skenario Bersamaan)...")
    Visualization.plot_flow_rates(simultaneous_sim)
    plt.show()
    
    # Plot volume untuk skenario bersamaan
    print("\n   Menampilkan grafik volume (Skenario Bersamaan)...")
    Visualization.plot_volume_profile(simultaneous_sim)
    plt.show()
    
    # 6. Analisis variasi parameter
    print("\n6. Analisis Variasi Parameter...")
    
    # Variasi kecepatan pompa
    pump_variations = ScenarioAnalysis.analyze_pump_variations(base_config)
    
    fig2, ax = plt.subplots(figsize=(10, 6))
    for var in pump_variations:
        ax.plot(var['simulator'].time_history, 
                var['simulator'].height_history,
                label=f"v = {var['velocity']} m/s")
    
    ax.set_xlabel('Waktu (detik)')
    ax.set_ylabel('Ketinggian (m)')
    ax.set_title('Pengaruh Kecepatan Pompa terhadap Waktu Pengisian')
    ax.legend()
    ax.grid(True, alpha=0.3)
    plt.show()
    
    # Tampilkan ringkasan
    print("\n" + "="*60)
    print("RINGKASAN HASIL SIMULASI")
    print("="*60)
    print("\nPertanyaan Studi Kasus:")
    print("1. Waktu pengisian hingga penuh:", 
          f"{filling_sim.results['time_to_full']:.1f} detik" if filling_sim.results['time_to_full'] else "Tidak tercapai")
    print("2. Waktu pengosongan hingga kosong:", 
          f"{emptying_sim.results['time_to_empty']:.1f} detik" if emptying_sim.results['time_to_empty'] else "Tidak tercapai")
    print("3. Profil ketinggian air telah ditampilkan dalam grafik")
    print("4. Pengisian & pengosongan bersamaan menghasilkan ketinggian akhir:", 
          f"{simultaneous_sim.results['final_height']:.2f} m")
    print("5. Ukuran tank optimal:", f"Tinggi {optimal['recommended_height']:.2f} m, Diameter {optimal['recommended_diameter']:.2f} m")
    print("="*60)
    print("SIMULASI SELESAI")
    print("="*60)

# Jalankan program
if __name__ == "__main__":
    main()
```
